```python
import pandas as pd
import matplotlib.pyplot as plt

# Specify the path to your CSV file
file_path = r"C:\Users\Paula Woest\OneDrive\Desktop\Bachelore Arbeit\seconds_info.csv"

# Read the CSV file
data = pd.read_csv(file_path)

# Display the first few rows of the DataFrame (optional)
print(data.head())

# Assuming your CSV has columns 'X' and 'Y' for the scatter plot
x_column = 's_in_roi'  # Name of your x-axis column
y_column = 's_in_roi_ang_dir'  # Name of your y-axis column

# Scatter plot to visualize the data
plt.figure(figsize=(10, 6))
plt.scatter(data[x_column], data[y_column], color='blue', marker='o')
plt.title('Scatter Plot of Your Data')
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.grid(True)
plt.show()

# Define the "corner" region where the data points cluster
# For example, assume you notice clustering in the lower-left corner of the plot
# Adjust the bounds based on your specific scatter plot
x_min = data[x_column].quantile(0.05)  # Bottom 5% for x
y_min = data[y_column].quantile(0.05)  # Bottom 5% for y

# Filter the data to get points in this "corner" region
corner_data = data[(data[x_column] <= x_min) & (data[y_column] <= y_min)]

# Show the filtered data (optional)
print("Data points in the corner:")
print(corner_data)

# Analyze the common features in the "corner" data
# For example, check the value counts for other columns
common_features = corner_data.describe(include='all')
print("\nCommon features in the corner data:")
print(common_features)

# Optional: Check specific columns for patterns
for col in data.columns:
    if col not in [x_column, y_column]:
        print(f"\nUnique values in column '{col}' for corner points:")
        print(corner_data[col].value_counts())
```
